## Problem: From the given ‘Iris’ dataset, Create the Decision Tree classifier and visualize it graphically.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.info()
```

```python
iris.describe().T
```

```python
data=iris.drop('Id',axis=1)
```

```python
import seaborn as sb
pairplot=sb.pairplot(data,hue='Species')
```

```python
x=data.iloc[:,[0,1,2,3]].values
y=data.iloc[:,[4]].values
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.2, random_state=10) 
```

### Define the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='gini', max_depth=5)
dtree.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

```python
y_pred=dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
```

```python
print(cm)
print(ac)
```

### visualize the Decision Tree to understand it better.


```python
from six import StringIO  
from IPython.display import Image  
from sklearn import tree
import pydotplus 
# Visualize the graph
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['Iris-setosa','Iris-versicolor','Iris-virginica']
plt.figure(figsize=(18,18))
tree.plot_tree(dtree,filled=True,feature_names=fn,class_names=cn)
```
